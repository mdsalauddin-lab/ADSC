```python
%%bash

echo "# ADSC" >> README.md
git init
git config user.email "you@example.com"
git config user.name "Your Name"
git add README.md
git commit -m "first commit"
git branch -M main

# Check if 'origin' remote exists. If not, add it. Otherwise, ensure URL is correct.
if ! git remote get-url origin > /dev/null 2>&1; then
    git remote add origin https://github.com/mdsalauddin-lab/ADSC.git
else
    # If origin exists, ensure it points to the correct URL (optional, can be omitted if URL is guaranteed).
    git remote set-url origin https://github.com/mdsalauddin-lab/ADSC.git
fi

# To push to GitHub, you need to authenticate using a Personal Access Token (PAT).
# Replace <YOUR_GITHUB_USERNAME> and <YOUR_PERSONAL_ACCESS_TOKEN> with your actual credentials.
# You can generate a PAT from GitHub's Developer settings (ensure 'repo' scope).
# WARNING: Including your PAT directly in the code is not recommended for security in production environments.
git push https://<YOUR_GITHUB_USERNAME>:<YOUR_PERSONAL_ACCESS_TOKEN>@github.com/mdsalauddin-lab/ADSC.git main
```
